```python
import pandas as pd

file_path = '/content/turkishMMLU.csv'

df = pd.read_csv(file_path)

df.head()

```

```python
df_cleaned = df[['soru', 'cevap', 'secenekler']]

df_cleaned.head()
```

```python
import ast

def fix_options_format(options):
    try:
        if isinstance(options, str):
            return ast.literal_eval(options)
    except (ValueError, SyntaxError):
        return options
    return options

df_cleaned['secenekler'] = df_cleaned['secenekler'].apply(fix_options_format)

df_cleaned.head()
```

```python
def map_answer_to_options(row):
    options = row['secenekler']
    answer_index = row['cevap']
    if isinstance(options, list) and 0 <= answer_index < len(options):
        return options[answer_index]
    else:
        return None

df_cleaned['dogru_cevap'] = df_cleaned.apply(map_answer_to_options, axis=1)

df_cleaned[['soru', 'dogru_cevap']].head()
```

```python
df_cleaned = df_cleaned[df_cleaned['dogru_cevap'].notnull()]

df_cleaned.shape
```

```python
df_cleaned = df_cleaned[['soru', 'dogru_cevap']]

df_cleaned.head()
```

```python
df_cleaned['input_text'] = df_cleaned.apply(lambda row: f"Soru: {row['soru']} Cevap: {row['dogru_cevap']}", axis=1)

df_cleaned[['input_text']].head()
```

```python
from datasets import Dataset

dataset = Dataset.from_pandas(df_cleaned[['input_text']])

dataset
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")
```

```python
df_cleaned['input_text'] = df_cleaned.apply(lambda row: f"Soru: {row['soru']}", axis=1)
df_cleaned['target_text'] = df_cleaned.apply(lambda row: f"Cevap: {row['dogru_cevap']}", axis=1)
from datasets import Dataset

dataset = Dataset.from_pandas(df_cleaned[['input_text', 'target_text']])

dataset
```

```python
def preprocess_data(examples):
    inputs = [f"{text}" for text in examples['input_text']]
    targets = [f"{text}" for text in examples['target_text']]

    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")
    labels = tokenizer(targets, max_length=512, truncation=True, padding="max_length").input_ids

    model_inputs["labels"] = labels
    return model_inputs

tokenized_dataset = dataset.map(preprocess_data, batched=True)
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    eval_dataset=tokenized_dataset,
)

trainer.train()
```
